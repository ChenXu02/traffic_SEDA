```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
data=np.loadtxt(open("SimMatrix.csv","rb"),delimiter=",",skiprows=0)  
data1=data[0:8]#original
data2=data[8:16]#PM20
data3=data[16:24]#PM40
data4=data[24:32]#CM20
data5=data[32:40]#CM40
x_ticks = ['P04_f','P04_o','P04_s','P08_f','P08_o','P08_s','CD_S','SZ_S']
y_ticks = ['P04_f','P04_o','P04_s','P08_f','P08_o','P08_s','CD_S','SZ_S']  
sns.set_context({"figure.figsize":(6,6)})
```

```python
sns.set_context({"figure.figsize":(6,6)})
ax = sns.heatmap(data1, xticklabels=x_ticks, yticklabels=y_ticks,square=True,vmin=0.75,vmax=1)
ax.set_title('Datasets Similarity (Original)',fontsize=20)  
ax.set_xlabel('Source Domain',fontsize=14)  
ax.set_ylabel('Target Domain',fontsize=14)
ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
ax.set_yticklabels(ax.get_yticklabels(),rotation=0)
figure = ax.get_figure()
figure.savefig('heatmap.png',dpi=300) 
```
