```python
import seaborn as sns
import numpy as np
data=np.loadtxt(open("SimMatrixoriginal.csv","rb"),delimiter=",",skiprows=0)  
data1=data[0:8]#original
data2=data[8:16]#PM20
data3=data[16:24]#PM40
data4=data[24:32]#CM40
data5=data[32:40]#CM20
x_ticks = ['P04_f','P04_o','P04_s','P08_f','P08_o','P08_s','CD_S','SZ_S']
y_ticks = ['P04_f','P04_o','P04_s','P08_f','P08_o','P08_s','CD_S','SZ_S']  
```

```python
ax = sns.heatmap(data1, xticklabels=x_ticks, yticklabels=y_ticks,square=True)
ax.set_title('Datasets Similarity (Original)')  
ax.set_xlabel('Source Domain')  
ax.set_ylabel('Target Domain')
figure = ax.get_figure()
figure.savefig('heatmap.jpg') 
```

```python
ax = sns.heatmap(data2, xticklabels=x_ticks, yticklabels=y_ticks,square=True)
ax.set_title('Datasets Similarity (PM 20%)')  
ax.set_xlabel('Source Domain')  
ax.set_ylabel('Target Domain')
figure = ax.get_figure()
figure.savefig('heatmap_pm20.jpg') 
```

```python
ax = sns.heatmap(data3, xticklabels=x_ticks, yticklabels=y_ticks,square=True)
ax.set_title('Datasets Similarity (PM 40%)')  
ax.set_xlabel('Source Domain')  
ax.set_ylabel('Target Domain')
figure = ax.get_figure()
figure.savefig('heatmap_pm40.jpg') 
```

```python
ax = sns.heatmap(data5, xticklabels=x_ticks, yticklabels=y_ticks,square=True)
ax.set_title('Datasets Similarity (CM 20%)') 
ax.set_xlabel('Source Domain')  
ax.set_ylabel('Target Domain')
figure = ax.get_figure()
figure.savefig('heatmap_cm20.jpg')  
```
