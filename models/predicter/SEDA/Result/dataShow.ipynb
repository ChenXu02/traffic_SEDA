```python
import numpy as np
import matplotlib.pyplot as plt
data=np.loadtxt(open('datashow2.csv'),delimiter=",",skiprows=0).T
print(data.shape)
```

```python
def lineChart(showData,m,d,s,scale):
    '''
    showData: input data
    m: missing type (PM, CM)
    d: dataset (PeMS08, CD_S)
    s: state of dataset (missing, imputed)
    scale: the scale of y-axis
    '''
    plt.gca().set_prop_cycle(plt.cycler('color', plt.cm.jet(np.linspace(0, 1, len(showData)))))# Plot several different functions..
    plt.grid()
    labels1 = []
    chart_title='MAPE of {}({},{})'.format(d,m,s)
    plt.xlabel('Missing Rate(%)', fontsize='12')  # name of x axis, size of word
    plt.ylabel('MAPE(%)', fontsize='12')  # name of y axis, size of word
    plt.ylim(8, scale)
    model=['GRU','DCRNN','AGCRN','BTTF','LSTM-GL-ReMF','SEDA']
    marker=['+','^','o','*','x','1']
    for i in range(0, len(showData) ):
        x, y = range(0, len(showData[0])*10,10), showData[i]  
        plt.plot(x, y,linewidth =2,linestyle='dashdot',marker=marker[i])
        labels1.append(model[i])
    plt.legend(labels1, ncol=2, loc='upper left',
               bbox_to_anchor=[0, 1],
           columnspacing=1.0, labelspacing=0.2,
           handletextpad=0.2, handlelength=1.5,
           fancybox=True, shadow=True)
    plt.savefig(chart_title+'.png',dpi=300)  #
    plt.show()
```

```python
missingtype=['PM','CM']
datasets=['PeMS_08','CD_S']
state=['Missing','Imputed']
```

```python
bias=0
scale=100
showData=data[:,0+bias*10:10+bias*10]
lineChart(showData,missingtype[0],datasets[0],state[0],scale)
```
