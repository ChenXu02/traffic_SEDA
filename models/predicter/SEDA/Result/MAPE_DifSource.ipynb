```python
import matplotlib.pyplot as plt
import numpy as np 

labels = ['P04(f)','P04(o)','P04(s)', 'SZ_S']
Original = [10.03, 10.04,9.91,9.89]
PM20 = [10.60, 10.58,10.03,9.99]
PM40 = [11.22, 11.17,10.29,10.13]
CM20 = [10.91, 10.96,10.15,10.02]
CM40 = [11.72, 11.66,10.47,10.29]
Sim_D=[0.8166,0.8073,0.9900,0.9897]
# marks = ["o", "X", "+", "*", "O"]

x = np.arange(len(labels))  
width = 0.16  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width * 2, Original, width, label='Original', hatch="...", color='#65ab7c', edgecolor="k")
rects2 = ax.bar(x - width + 0.01, PM20, width, label='PM20%', hatch="oo", color='#6140ef', edgecolor="k")
rects3 = ax.bar(x + 0.02, PM40, width, label='PM40%', hatch="++", color='#fd8d49', edgecolor="k")
rects4 = ax.bar(x + width + 0.03, CM20, width, label='CM20%', hatch="XX", color='#fbeeac', edgecolor="k")
rects5 = ax.bar(x + width * 2 + 0.04, CM40, width, label='CM40%', hatch="**", color='#ce5dae', edgecolor="k")
ax.yaxis.grid(color='b',linestyle='--',linewidth=1,alpha=0.3)
ax.set_ylabel('MAPE', fontsize=12)
ax.set_xlabel('Source', fontsize=12)
ax.set_xticks(x)
ax.set_ylim(ymin = 9, ymax = 12)
ax.set_xticklabels(labels)
ax2 = ax.twinx()

for a, b in zip(x[0:2], Sim_D[0:2]):
    plt.text(a, b+0.01, '%.4f'%b, ha='center', va='bottom', fontsize=12)
for a, b in zip(x[2:], Sim_D[2:]):
    plt.text(a, b-0.05, '%.4f'%b, ha='center', va='bottom', fontsize=12)

ax2.plot(x, Sim_D,'b^--', color="red", label="Sim_D")
ax2.set_ylabel("Sim_D", fontsize=12)
ax2.set_ylim(ymin = 0.4, ymax = 1)

fig.legend(loc="upper right", bbox_to_anchor=(1, 0.93), bbox_transform=ax.transAxes,prop={'size':11})
plt.savefig('MAPE_DifSource.png',dpi=600)
plt.show()
```
